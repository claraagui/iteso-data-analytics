### Probabilidad condicional
#
Con el histograma anterior podemos observar que la probabilidad de que sobrevivieron dado que son mujeres aproximadamente de 74%, la probabilidad de que sobrevivan dado a que son hombres es de aproximadamente de 18%.

```python
df_pro_hombres['Age'].min(),df_pro_hombres['Age'].max()
```

```python
df_pro_mujeres['Age'].min(),df_pro_mujeres['Age'].max()
```

```python
edad=0
for i in range(10,90,10):
    #print(edad,i)
    personas_hombres = df_pro_hombres[(df_pro_hombres['Age']>=edad)&(df_pro_hombres['Age']<i)]
    plt.figure()
    n=plt.hist(personas_hombres['Survived'], density=True)
    plt.title('hombres survived en el rango de edad:'+str(edad)+','+str(i))
    edad=i
    print(n)
```

```python
edad=0
for i in range(10,80,10):
    #print(edad,i)
    personas_mujeres = df_pro_mujeres[(df_pro_mujeres['Age']>=edad)&(df_pro_mujeres['Age']<i)]
    plt.figure()
    n= plt.hist(personas_mujeres['Survived'], density=True)
    plt.title('mujeres survived en el rango de edad:'+str(edad)+','+str(i))
    edad=i
    print(n)
```

La probabilidad de que una persona siendo hombre haya sobrevivido dado que su edad estÃ¡ en el rango de 0 a 9 aÃ±os es del 59% aproximadamente, en el rango de 10 a 19 serÃ­a de 12%, en el rango de 20 a 29 serÃ­a de 15%, el rango de 30 a 39 serÃ­a del 21%, el rango de 40 a 49 aÃ±os es de 21%, en el rango de 50 a 59 aÃ±os es de 13%, en el rango de 60 a 69 es del 13%, en el rango de 70 a 79 0% de sobrevivientes.

La probabilidad de que una persona siendo mujer haya sobrevivido dado que su edad estÃ¡ en el rango de 0 a 9 aÃ±os es del 63% aproximadamente, en el rango de 10 a 19 serÃ­a de 75%, en el rango de 20 a 29 serÃ­a de 70%, el rango de 30 a 39 serÃ­a del 83%, el rango de 40 a 49 aÃ±os es de 68%, en el rango de 50 a 59 aÃ±os es de 88%, en el rango de 60 a 69 es del 100%.
