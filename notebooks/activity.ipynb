Aplicando el algoritmo de K-Means, con random-init

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=5, init='random')
model= model.fit(df_result_kmeans)
grupos = model.predict(df_result_kmeans)
centroids=model.cluster_centers_
```

```python
df_result_kmeans['labels_kmeans']=grupos
```

```python
unique_labels = set(grupos)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    xy = df_result_kmeans[df_result_kmeans['labels_kmeans']==k].values
    plt.plot(xy[:, 3],xy[:, 1],"o",markerfacecolor=tuple(col),markeredgecolor="k",markersize=14,)
plt.title(f"K-Means clustering")
plt.show()
```

Aplicando el algoritmo de K-Means++

```python
model=KMeans(n_clusters=5, init='k-means++')
model= model.fit(df_result_kmean_plus_plus)
grupos = model.predict(df_result_kmean_plus_plus)
centroids=model.cluster_centers_
```

```python
df_result_kmean_plus_plus['labels_kmeans++']=grupos
```

```python
unique_labels = set(grupos)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    xy = df_result_kmean_plus_plus[df_result_kmean_plus_plus['labels_kmeans++']==k].values
    plt.plot(xy[:, 2],xy[:, 1],"o",markerfacecolor=tuple(col),markeredgecolor="k",markersize=14,)
plt.title(f"K-Means++ clustering")
plt.show()
```

Aplicando el DBSCAN

```python
df_result
```

```python
df_result.set_index('host_id', inplace=True)
```

```python
df_result
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
df_result_scaler=scaler.fit_transform(df_result)  #fit entrenar al algoritmo, se realiza la estandarización de los datos
```

```python
df_result_scaler
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.5, min_samples=5).fit(df_result_scaler)
labels=db.labels_
```

```python
labels
```

```python
set(labels)
```

```python
# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print("Clusters estimados por DBSCAN: %d" % n_clusters_)
print("Numeros estimatos de muestras con ruido: %d" % n_noise_)
```

```python
df_result['labels_dbscan']=labels
```

```python
df_result_dbscan=df_result[df_result['labels_dbscan']!=-1]  #muchos valores de ruido (-1), se ploteo porque eso da dbscan
```

```python
df_result_dbscan
```

```python
labels_dbscan=df_result_dbscan['labels_dbscan'].values
```

```python
set(labels_dbscan)  #set sirve para crear los conjuntos y saber los valores que hay en este arreglo, 
                    #y los valores únicos que se obtienen son los obtenidos por dbscan.
```

```python
unique_labels = set(labels_dbscan)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:# Black used for noise.
        col = [0, 0, 0, 1]
    class_member_mask = labels_dbscan == k
    xy = df_result_dbscan[df_result_dbscan['labels_dbscan']==k].values
    plt.plot(xy[:, 2],xy[:, 7],"o",markerfacecolor=tuple(col),markeredgecolor="k",markersize=14,)
plt.title(f"Estimated number of clusters: {n_clusters_}")
plt.show()
```
