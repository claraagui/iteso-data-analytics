```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv("../data/raw/train.csv")
df
```

```python
df['Family_size'] = df.SibSp + df.Parch+1
df['LastName'] = df['Name'].apply(lambda x: x.split(',')[0].strip())
df
```

```python
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked','Family_size','LastName']]
y = df['Survived']

X.isnull().sum()
```

```python
#Reemplazar datos faltantes
X['Age'].fillna(X['Age'].median(), inplace=True)
X['Fare'].fillna(X['Fare'].median(), inplace=True)
X['Embarked'].fillna(X['Embarked'].mode()[0], inplace=True)  


X = pd.get_dummies(X, columns=['Sex', 'Embarked','LastName'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Estandarizar los features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Modelo KNN
knn = KNeighborsClassifier(metric='cosine')
knn.fit(X_train, y_train)

# Predecir la y
y_pred = knn.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Modelo KNN
knn = KNeighborsClassifier(metric='euclidean')
knn.fit(X_train, y_train)

# Predecir la y
y_pred = knn.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
