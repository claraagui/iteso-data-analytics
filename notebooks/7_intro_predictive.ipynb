```python
from sklearn.datasets import make_regression


X,y= make_regression(n_features=1, noise=40)
```

```python
X.shape, y.shape
```

```python
import pandas as pd 

df=pd.DataFrame(X)
df.columns=['X']
df['y']=y
df
```

```python
df['X'].plot(kind='hist', color='lavender')
```

```python
import numpy as np
 
y_mean=np.array([y.mean()]*100)
y_mean
```

```python
from sklearn.linear_model import LinearRegression

model1= LinearRegression() 
model2= LinearRegression() 
model3= LinearRegression()  
#modelo fit  es donde entrenamos a un modelo
model1.fit(X,y)

# una clase es lo que define a un comportamiento
# un objeto es la instancia de una clase
```

```python
y_pred=model1.predict(X)
```

```python
import matplotlib.pyplot as plt

X,y= make_regression(n_features=1,noise=30, random_state=0)

# Selector
plt.figure(figsize=(10,4))

# Data
plt.scatter(X,y, label='Random Data')

plt.plot(X, y_mean, label='Mean',c='green')

plt.plot(X, y_pred, label='Linear Regression',c='red')

# Show
plt.xlabel('Features (X)')
plt.ylabel('Target (y)')
plt.title('Scatter Example Plot')
plt.legend()

plt.savefig("../reports/figures/random1.png")
```
